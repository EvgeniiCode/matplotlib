Построим различные виды графиков

```python
plt.plot([1, 2, 3], [1, 2, 3])
```

Построим треугольник, последними точками замыкаем его контур

```python
plt.plot((1, 5, 10, 1), (1, 10, 1, 1))
```

Построим точечный график

```python
plt.scatter((2, 4, 8), [2, 4, 8])
```

Построим bar график

```python
plt.bar(['apple', 'orange', 'banana'], (10, 25, 20))
```

Построим горизонтальный график

```python
plt.barh(('apple', 'banana', 'orange'), [25, 10, 30])
```

### Интересное применение графика bar, главнео задать ширину столбцов и учесть расстояние изначальное между ними по оси ox 

```python
plt.bar([5.8, 6.8, 7.8], [10, 15, 21], width=0.2)
plt.bar([6, 7, 8], [6, 12, 20], width=0.2)
plt.show()
```

Построим круговую диаграмму. Количество элементов в массиве определяет количество клиньев, а величина значений определяет их площадь

```python
plt.pie([10, 20 , 30, 20 , 20])
plt.show()
```

Построим график с усами

```python
plt.boxplot([[1, 5, 7, 4, 6, 10, 15],
             [-2, 5, 7, 4, 6, 10, 15],
             [-4, 5, 7, 4, 6, 10]])
plt.show()
```
