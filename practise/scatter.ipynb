### Рассмотрим график scatter. Для данного графика свои параметры настройки(см документацию). Так же можно делать фигуру больше или точки особым маркерм для наибольшей наглядности

```python
import matplotlib as mb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.random.rand(5000)
y1 = np.random.gamma(1, size = 5000)
y2 = np.random.gamma(2, size = 5000)
y3 = np.random.gamma(4, size = 5000)
y4 = np.random.gamma(8, size = 5000)
```

```python
fig, ax = plt.subplots()
```

```python
fig.set(figwidth=20, figheight=20)
```

```python
ax.scatter(x, y1,
           color = 'blue',
           s = 1, alpha=0.8, marker='*')
# RGB:
ax.scatter(x + 1, y2,
           c = [[0.1, 0.63, 0.55]],
           s = 1)
# hex RGB:
ax.scatter(x + 2, y3,
           c = '#ad09a3',
           s = 1)
# уровень серого в интервале [0, 1]:
ax.scatter(x + 3, y4,
           c = ['0.9'],
           s = 1)
```
