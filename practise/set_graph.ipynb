### Рассмотрим параметры графиков и их построение

```python
import matplotlib as mb
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
```

```python
x = [int(i) for i in range(10)]
y = [i ** 2 for i in x]
x1 = x
y1 = [i ** 3 for i in x1]
```

```python
ax.plot(x, y, color='red')
ax.plot(x1, y1, color='green', linewidth=1)
plt.show()
```
