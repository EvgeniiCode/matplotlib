## Рассмотрим backend для matplotlib

Бывают различные версии бэкенда для отображения графиков

```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
print(matplotlib.__version__)
```

### %matplotlib inline это первый backend это стандартный для jupyter стандартный 

```python
%matplotlib inline 
```

```python
print(matplotlib.get_backend())
```

```python
plt.plot((1, 2, 3), (1, 2, 3))
```

### второй backend это Qt5Agg %matplotlib. Он позволяет на одном графике в интерактивном режиме отображать все данные. Каждый раз добавляя новые и не удаля старые

```python
%matplotlib
```

```python
matplotlib.use('Qt5Agg')
```

```python
matplotlib.get_backend()
```

```python
plt.plot((1, 2, 3), (1, 2, 3))
```

```python
plt.scatter((1, 2, 3), (4 , 5, 6))
```

```python
plt.bar(['a', 'b'], (2 , 2))
```

### третий backend  это %matplotlib notebook. Показывает все изменения на одном графике в интерактивном режиме.

```python
%matplotlib notebook
```

```python
matplotlib.use('nbagg')
```

```python
matplotlib.get_backend()
```

```python
plt.plot((1, 2, 3))
```

```python
plt.plot((1, 5), (7, 2))
plt.show()
```

```python
plt.scatter((1, 2, 3), (1, 2, 3))
```
